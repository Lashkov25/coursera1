**Перекодування категоріальних даних**

```python
from sklearn.preprocessing import OrdinalEncoder

# Отримати унікальні значення для кожного з цих стовпців
unique_countries = df['Country'].unique().tolist()
unique_regions = df['Region'].unique().tolist()
unique_income_groups = df['Income group'].unique().tolist()
unique_political_regimes = df['Political regime'].unique().tolist()

# Ініціалізуємо OrdinalEncoder з власним порядком
encoder_country = OrdinalEncoder(categories=[unique_countries])
encoder_region = OrdinalEncoder(categories=[unique_regions])
encoder_income_group = OrdinalEncoder(categories=[unique_income_groups])
encoder_political_regime = OrdinalEncoder(categories=[unique_political_regimes])

# Перекодовуємо дві категоріальні колонки
df['Country'] = encoder_country.fit_transform(df[['Country']])
df['Region'] = encoder_region.fit_transform(df[['Region']])
df['Income group'] = encoder_income_group.fit_transform(df[['Income group']])
df['Political regime'] = encoder_political_regime.fit_transform(df[['Political regime']])

# Перевіряємо результат
print(df)
```

**Стандартизація даних**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # обираємо бажаний метод стандартизації
scaler.fit(df)
X_stand = scaler.fit_transform(df)
```

# Завдання 3

```python
# реалізація методу зігнутого ліктя
from sklearn.cluster import KMeans
klusters = range(2, 11) # задамо кількість кластерів від 2 до 10
inertias = []

for k in klusters:
    # задаємо метод кластеризації - к-середніх
    model = KMeans(n_clusters=k, random_state=42)

    # оцінюємо модель
    model.fit(X_stand)

    # для кожної к-ті кластерів розраховуємо inertia, розраховані значення додаємо в список inertias
    inertias.append(model.inertia_)

# зобразимо на графіку показник inertia для кожної к-ті кластерів
plt.figure(figsize=(7,4))
plt.plot(klusters, inertias, '-o', color='red')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
plt.xticks(klusters)
plt.show()
```

```python
# Розбиваємо дані на 5 кластерів
kmeans = KMeans(n_clusters=5, random_state=42)
#random_state дозволяє "закріпити" присвоєння кожній групі номеру кластеру при багаторазовому повторенні коду

kmeans.fit(X_stand)

# Отримуємо відповідні кластери
labels = kmeans.predict(X_stand)
print(labels)
```

```python
# додамо отримані значення кластерів до таблиці із початковими даними
df['Cluster'] = labels
df.head()
```

```python
#перевіримо наповненість кластерів
df[['Cluster']].value_counts()
```

```python
centers = kmeans.cluster_centers_
```

```python
# додамо отримані центри кластерів у датафрейм new_df
new_df = pd.DataFrame(centers, columns = [['Country','Talent','Infrastructure','Operating Environment','Research','Development',
	                                          'Government Strategy','Commercial',	'Total score','Region','Income group','Political regime']])
new_df
```

```python
# Середні за кластерами значення
mean_values = df.groupby('Cluster').mean()
print(mean_values)
```

```python
# Підрахуємо, які країни входять в кожен кластер
clustered_countries = df.groupby('Cluster')['Country'].apply(list)
print(clustered_countries)
```
